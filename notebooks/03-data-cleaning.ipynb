## Carga y Limpieza de Datos

```python
from pathlib import Path

import polars as pl
from loguru import logger
```

```python
DATA_DIR = Path("../data/raw")

def update_consolidate(folder: str, format: str = "parquet"):

    logger.info(f'Cargando datos desde {folder} con formato {format}...')
    return pl.scan_parquet(f"{folder}/*.{format}").collect()

all_dataset = update_consolidate(DATA_DIR, format="parquet")
all_dataset.head(5).show()
```

## Data Cleaning basado en el EDA

```python
drop_columns = [
    # Evitar leakage de información, eliminar columnas que no estarán disponibles en el momento de la predicción
    'punt_ingles',
    'punt_matematicas',
    'punt_lectura_critica',
    'punt_sociales_ciudadanas',
    'punt_c_naturales',
    'desemp_ingles',

    # Identificadores (sin señal generalizable)
    'estu_consecutivo',
    'cole_cod_dane_establecimiento',
    'cole_cod_dane_sede',
    'cole_codigo_icfes',
    'cole_nombre_establecimiento',
    'cole_nombre_sede',

    'estu_cod_depto_presentacion',
    'estu_cod_mcpio_presentacion',
    'estu_cod_reside_depto',
    'estu_cod_reside_mcpio',
    'cole_cod_depto_ubicacion',
    'cole_cod_mcpio_ubicacion',

    # Prácticamente constantes, no aportan varianza
    'estu_estudiante',
    'estu_privado_libertad',
    'estu_estadoinvestigacion',
    'estu_nacionalidad',
    'estu_pais_reside'
]

print(f"Columnas originales: {len(all_dataset.columns)}")
print(f"Columnas eliminadas: {len(drop_columns)}")
print(f"Columnas restantes: {len(all_dataset.columns) - len(drop_columns)}")
```

```python
cleaned_dataset = (
    all_dataset.clone()
    .with_columns(
        pl.col('estu_consecutivo')
        .cum_count()
        .over('estu_consecutivo')
        .alias('duplicate_count')
    )
    .filter(pl.col('duplicate_count') == 1)
    .drop(drop_columns + ['duplicate_count'])
    .filter(pl.col("punt_global").is_not_null())
)

print(f"Columnas después de limpieza: {len(cleaned_dataset.columns)}")
print(f"Dataset después de eliminar filas duplicadas: {cleaned_dataset.shape[0]:,} filas restantes")
```

```python
DATA_DIR_PROCESSED = Path("../data/processed")
cleaned_dataset.write_parquet(DATA_DIR_PROCESSED / "cleaned_dataset.parquet")
```
